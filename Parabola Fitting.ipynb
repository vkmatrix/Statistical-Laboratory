```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#open upward
x=np.linspace(-1,1,30)
y=x**2
plt.plot(x,y);
```

```python
#open downward
x=np.linspace(-1,1,30)
y=-x**2
plt.plot(x,y);
```

```python
#open rightward
y=np.linspace(-1,1,30)
x=y**2
plt.plot(x,y);
```

```python
#open leftward
y=np.linspace(-1,1,30)
x=-y**2
plt.plot(x,y);
```
