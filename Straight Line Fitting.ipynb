```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#horizontal line
x=np.linspace(-1,1,3)
y=[5,5,5]
plt.plot(x,y,'y',color='r',linestyle='--',linewidth=3)
```

```python
#vertical line
x=[5,5,5]
y=np.linspace(-1,1,3)
plt.plot(x,y,'--g');
```

```python
#line with positive slope
x=np.linspace(-1,1,3)
y=x
plt.plot(x,y,'b--',linewidth=3);
```

```python
#line with negative slope
x=np.linspace(-1,1,3)
y=-x
plt.plot(x,y,'y--',linewidth=3);
```
