```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#simple bar plot
courses=['cse','it','mech','ece','ai&ds']
values=[180,180,120,180,120]
fig=plt.figure(figsize=(10,6))
plt.bar(courses,values,color=['green','blue','yellow','cyan','tan'],width=0.5);
```

```python
#unstacked bar plot
import pandas as pd
df=pd.DataFrame({"2018":[57,67,77,83],"2019":[68,73,80,79],"2020":[73,78,80,85]},index=['A','B','C','D'])
df.plot(kind='bar',stacked=False,figsize=(10,5),color=['blue','green','red','black'])
plt.title('Goal Report of Each Player from 2018 to 2020')
plt.xlabel('Players')
plt.ylabel('Goals Scored')
plt.show()
```

```python
#unstacked bar plot
import pandas as pd
df=pd.DataFrame({"2018":[57,67,77,83],"2019":[68,73,80,79],"2020":[73,78,80,85]},index=['A','B','C','D'])
df.plot(kind='bar',stacked=True,figsize=(10,5),color=['blue','green','red','black'])
plt.title('Goal Report of Each Player from 2018 to 2020')
plt.xlabel('Players')
plt.ylabel('Goals Scored')
plt.show()
```
