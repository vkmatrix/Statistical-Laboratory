```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
grades=['S','A','B','C','D','E','F']
values=[5,15,20,5,2,2,1]
```

```python
#simple pie plot
fig=plt.figure(figsize=(10,6))
plt.pie(values,labels=grades)
plt.title('Simple Pie Plot')
plt.show()
```

```python
# explode pie plot
fig=plt.figure(figsize=(10,6))
exp=[0,0,0,0,0,0,0.6]
plt.pie(values,labels=grades,explode=exp);
plt.title('exploded pie chart');
```

```python
#exploded with shadow pie chart
fig=plt.figure(figsize=(10,6))
plt.pie(values,labels=grades,explode=exp,shadow=True,startangle=180,autopct='%2.1f%%')
plt.title('Explode with Shadow Pie Chart');
```

```python
#exploded shadow pie chart with customized colors and plot legends
fig=plt.figure(figsize=(10,8))
plt.pie(values,labels=grades,startangle=180,autopct="%2.1f%%",explode=exp,shadow=True,colors=['brown','cyan','yellow','blue','green','orange','red']);
plt.title('Performance Analysis of Students by means of their grades using Customized Exploded Shadow Pie Chart')
plt.legend(title='Grades');
```
